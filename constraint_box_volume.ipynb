```python
import numpy as np
from scipy.optimize import minimize
```

```python
def calcVolume(x):
  length = x[0]
  width = x[1]
  height = x[2]
  volume = length * width * height
  return volume
```

```python
def calcSurface(x):
  length = x[0]
  width = x[1]
  height = x[2]
  surfaceArea = 2 *length * width + 2 * width * height + 2 * width *  length
  return surfaceArea
```

```python
def objective(x):
  return -calcVolume(x)
```

```python
def constraint(x):
  return 10 - calcSurface(x)
```

```python
cons = ({'type': 'ineq', 'fun': constraint})
```

```python
lengthGuess = 1
widthGuess = 1
heightGuess = 1
```

```python
x0 = np.array([lengthGuess, widthGuess, heightGuess])
```

```python
sol = minimize(objective, x0, method='SLSQP', constraints = cons, options={'disp': True})
```

```python
print(sol)
```

```python
xOpt = sol.x
volumeOpt = -sol.fun
print('xOpt:',xOpt)
print('volumeOpt:',volumeOpt)
```

```python
surfaceAreaOpt = -sol.fun
```

```python
surfaceAreaOpt = calcSurface(xOpt)
```

```python
print('Length:' + str(xOpt[0]))
print('width:' + str(xOpt[1]))
print('height:' + str(xOpt[2]))
print('Volume:' + str(volumeOpt))
print('Surface Area:' + str(surfaceAreaOpt))
```
