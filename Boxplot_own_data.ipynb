```python
# import pandas
import pandas as pd
# import matplotlib
import matplotlib.pyplot as plt
# import seaborn
import seaborn as sns
%matplotlib inline
```

```python
#data_url = 'http://bit.ly/2cLzoxH'
data_url = 'https://raw.githubusercontent.com/abirhazra/Basic_Python/main/data.csv'
# read data from url as pandas dataframe
gapminder = pd.read_csv(data_url)
print(gapminder.head(10))
```

```python
gapminder_2007 = gapminder[gapminder['ID']==0]
gapminder_2007.shape
print(gapminder_2007)
```

```python
gapminder_2007.boxplot(by='FailureRate', 
                       column=['Device'], 
                       grid=False)
```

```python
bplot = sns.boxplot(y='FailureRate', x='Device', 
                 data=gapminder_2007, 
                 width=0.5,
                 palette="colorblind")
```

```python
# make boxplot with Seaborn
bplot=sns.boxplot(y='FailureRate', x='Device', 
                 data=gapminder_2007, 
                 width=0.5,
                 palette="colorblind")
 
# add stripplot to boxplot with Seaborn
bplot=sns.stripplot(y='FailureRate', x='Device', 
                   data=gapminder_2007, 
                   jitter=True, 
                   marker='o', 
                   alpha=0.5,
                   color='black',)
```

```python
# plot boxplot with seaborn
bplot=sns.boxplot(y='FailureRate', x='Device', 
                 data=gapminder_2007, 
                 width=0.5,
                 palette="colorblind")
 
# add swarmplot
bplot=sns.swarmplot(y='FailureRate', x='Device',
              data=gapminder_2007, 
              color='black',
              alpha=0.75)
```

```python
#bplot.axes.set_title("2007: Life Expectancy Vs Continent",fontsize=15)
 
bplot.set_xlabel("Devices", 
                fontsize=15)
 
bplot.set_ylabel("Failure Rate",
                fontsize=15)
 
bplot.tick_params(labelsize=15)
```

```python
# output file name
plot_file_name="boxplot_and_swarmplot_with_seaborn.pdf"
 
# save as jpeg
bplot.figure.savefig(plot_file_name,
                    format='pdf',
                    dpi=100)
```
