```python
import numpy as np
from scipy.optimize import minimize
```

```python
def objective_function(x):
  x1 = x[0]
  x2 = x[1]
  return x1**2 + x1*x2
  
```

```python
def equality_constraint(x):
  x1 = x[0]
  x2 = x[1]
  return x1**3 + x1*x2 - 100
  
```

```python
def inequality_constraint(x):
  x1 = x[0]
  x2 = x[1]
  return x1**3 + x1*x2 - 50
  
```

```python
bounds_x1 = (-100,100)
bounds_x2 = (-100,100)
```

```python
bounds = [bounds_x1, bounds_x2]
bounds
```

```python
constraint1 = {'type': 'eq', 'fun': equality_constraint}
constraint2 = {'type': 'eq', 'fun': inequality_constraint}
```

```python
constraint = [constraint1,constraint2]
```

```python
x0 = [1,1]
```

```python
result = minimize(objective_function, x0, method = 'SLSQP', bounds = bounds, constraints = constraint)
print(result)
```
