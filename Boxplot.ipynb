```python
# import pandas
import pandas as pd
# import matplotlib
import matplotlib.pyplot as plt
# import seaborn
import seaborn as sns
%matplotlib inline
```

```python
data_url = 'https://github.com/abirhazra/Basic_Python/blob/424ea5f8f8ccac585446b2579e065accac9c095a/data.csv'
# read data from url as pandas dataframe
gapminder = pd.read_csv(data_url)
print(gapminder.head(2))
```

```python
gapminder_2007 = gapminder[gapminder['year']==2007]
gapminder_2007.shape
```

```python
gapminder_2007.boxplot(by='continent', 
                       column=['lifeExp'], 
                       grid=False)
```

```python
bplot = sns.boxplot(y='lifeExp', x='continent', 
                 data=gapminder_2007, 
                 width=0.5,
                 palette="colorblind")
```

```python
# make boxplot with Seaborn
bplot=sns.boxplot(y='lifeExp', x='continent', 
                 data=gapminder_2007, 
                 width=0.5,
                 palette="colorblind")
 
# add stripplot to boxplot with Seaborn
bplot=sns.stripplot(y='lifeExp', x='continent', 
                   data=gapminder_2007, 
                   jitter=True, 
                   marker='o', 
                   alpha=0.5,
                   color='black')
```
