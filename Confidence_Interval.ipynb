```python
import numpy as np
from scipy import stats

# Sample data (replace with your dataset)
data = [85, 88, 92, 78, 95, 89, 91, 82, 79, 93]

# Confidence level (e.g., 95% confidence interval)
confidence_level = 0.95

# Calculate the mean and standard error of the sample
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)  # Use ddof=1 for sample standard deviation

# Calculate the margin of error
z_score = stats.norm.ppf((1 + confidence_level) / 2)  # Z-score for a two-tailed test
margin_of_error = z_score * (sample_std / np.sqrt(len(data)))

# Calculate the confidence interval
lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

# Print the results
print(f"Sample Mean: {sample_mean}")
print(f"Margin of Error: {margin_of_error}")
print(f"Confidence Interval ({100 * confidence_level}%): ({lower_bound}, {upper_bound})")
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data (replace with your data)
data = [85, 88, 92, 78, 95, 89, 91, 82, 79, 93]

# Confidence interval values (replace with your values)
lower_bound = 82
upper_bound = 91

# Create a ridge plot
sns.set(style="whitegrid")
sns.kdeplot(data, color="b", label="Data")

# Add vertical line for the mean
plt.axvline(np.mean(data), color='r', linestyle='--', label='Mean')

# Add shaded region for the confidence interval
plt.fill_betweenx([0], lower_bound, upper_bound, color='g', alpha=0.3, label='Confidence Interval')

# Set labels and title
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Ridge Plot with Confidence Interval')

# Show legend
plt.legend()

# Show the plot
plt.show()
```
